```python
import numpy as np
from ParseMetadata import ParseMetadataTree
```

```python
md = ParseMetadataTree('/home/prayush/src/SimulationAnnex/')
```

```python
SEPS = 1e-3
EPS = 1e-2
```

```python
for sim in md:
    sim_md = md[sim]
    for lcnt, lev in enumerate(sim_md):
        # Process only one lev for each
        if lcnt > 0: continue
        
        # Get q and spins for this simulation
        sim_lev_md = sim_md[lev]
        try: s1split = sim_lev_md['initial-dimensionless-spin1'].split()
        except:
            try: s1split = sim_lev_md['initial-spin1'].split()
            except: s1split = sim_lev_md['relaxed-spin1'].split()
        try: s2split = sim_lev_md['initial-dimensionless-spin2'].split()
        except:
            try: s2split = sim_lev_md['initial-spin2'].split()
            except: s2split = sim_lev_md['relaxed-spin2'].split()
        
        sim_s1x, sim_s1y, sim_s1z = np.array([float(val.strip(',')) for val in s1split])
        sim_s2x, sim_s2y, sim_s2z = np.array([float(val.strip(',')) for val in s2split])
        sim_q = float(sim_lev_md['initial-mass1']) / float(sim_lev_md['initial-mass2'])
        #print sim_q, sim_s1x, sim_s1y, sim_s1z, sim_s2x, sim_s2y, sim_s2z
        
        # Check if this simulations is precessing or not?
        if abs(sim_s1x) < SEPS and abs(sim_s1y) < SEPS and abs(sim_s2x) < SEPS and abs(sim_s2y) < SEPS:
            #print "Simulation %s is NOT precessing .. SKIPPING IT." % sim
            continue
        
        # Loop over all other simulations and see which match.
        for other_sim in md:
            if other_sim == sim: continue
            other_sim_md = md[other_sim]
            for other_lcnt, other_lev in enumerate(other_sim_md):
                if other_lcnt > 0: continue
                
                # Get q and spins for this simulation
                other_sim_lev_md = other_sim_md[other_lev]
                try: os1split = other_sim_lev_md['initial-dimensionless-spin1'].split()
                except:
                    try: os1split = other_sim_lev_md['initial-spin1'].split()
                    except:
                        try: os1split = other_sim_lev_md['relaxed-spin1'].split()
                        except: print "FOUND NO SPIN1 for %s amongst" % other_sim, other_sim_lev_md.keys()
                try: os2split = other_sim_lev_md['initial-dimensionless-spin2'].split()
                except:
                    try: os2split = other_sim_lev_md['initial-spin2'].split()
                    except: os2split = other_sim_lev_md['relaxed-spin2'].split()
        
                osim_s1x, osim_s1y, osim_s1z = np.array([float(val.strip(',')) for val in os1split])
                osim_s2x, osim_s2y, osim_s2z = np.array([float(val.strip(',')) for val in os2split])
                osim_q = float(other_sim_lev_md['initial-mass1']) / float(other_sim_lev_md['initial-mass2'])
                
                # Match both simulations:
                if abs(sim_q - osim_q) < EPS and abs(sim_s1z - osim_s1z) < SEPS and abs(sim_s2z - osim_s2z) < SEPS:
                    if abs(osim_s1x) < SEPS and abs(osim_s1y) < SEPS and abs(osim_s2x) < SEPS and abs(osim_s2y) < SEPS and abs(osim_s1z) > SEPS and abs(osim_s2z) > SEPS:
                        print "\n Sims %s and %s match our criteria!" % (sim, other_sim)
                        print "spins of first = ", sim_s1x, sim_s1y, sim_s1z, sim_s2x, sim_s2y, sim_s2z
                        print "spins of second = ", osim_s1x, osim_s1y, osim_s1z, osim_s2x, osim_s2y, osim_s2z
                        if abs(osim_s1z) < SEPS and abs(osim_s2z) < SEPS:
                            print "..second is NONSPINNING THOUGH.."
                        else: print ".. PERFECT MATCH!"
                        print "\n"
```

```python
for sim in md:
    sim_md = md[sim]
    for lcnt, lev in enumerate(sim_md):
        # Process only one lev for each
        if lcnt > 0: continue
        
        # Get q and spins for this simulation
        sim_lev_md = sim_md[lev]
        try: s1split = sim_lev_md['initial-dimensionless-spin1'].split()
        except:
            try: s1split = sim_lev_md['initial-spin1'].split()
            except: s1split = sim_lev_md['relaxed-spin1'].split()
        try: s2split = sim_lev_md['initial-dimensionless-spin2'].split()
        except:
            try: s2split = sim_lev_md['initial-spin2'].split()
            except: s2split = sim_lev_md['relaxed-spin2'].split()
        
        sim_s1x, sim_s1y, sim_s1z = np.array([float(val.strip(',')) for val in s1split])
        sim_s2x, sim_s2y, sim_s2z = np.array([float(val.strip(',')) for val in s2split])
        sim_q = float(sim_lev_md['initial-mass1']) / float(sim_lev_md['initial-mass2'])
        #print sim_q, sim_s1x, sim_s1y, sim_s1z, sim_s2x, sim_s2y, sim_s2z
        
        # Check if this simulations is precessing or not?
        if abs(sim_s1x) < SEPS and abs(sim_s1y) < SEPS and abs(sim_s2x) < SEPS and abs(sim_s2y) < SEPS:
            #print "Simulation %s is NOT precessing .. SKIPPING IT." % sim
            continue
        
        # Make sure NS spin = 0
        if abs(sim_s2x) > SEPS or abs(sim_s2y) > SEPS or abs(sim_s2z) > SEPS:
            continue
        
        # Exclude a range of mass ratios
        q = sim_q
        q = max(q, 1./q)
        if q < 3.0 or q > 5.0:
            continue
        print "Simulation: ", sim_lev_md['alternative-names']
        print "mass ratio = ", q
        print "spins of first = %.6f,%.6f,%.6f,\n\t second = %.6f,%.6f,%.6f" %\
                            (sim_s1x, sim_s1y, sim_s1z, sim_s2x, sim_s2y, sim_s2z)
        print "starting radial separation = ", sim_lev_md['initial-separation']
        print "number of orbits = ", sim_lev_md['number-of-orbits']
        print "time to merger = ", sim_lev_md['common-horizon-time']
        print "PERFECT MATCH!\n\n"
```

```python
ic1, ic2, ic3, ic4 = 0,0,0,0

for sim in md:
    sim_md = md[sim]
    for lcnt, lev in enumerate(sim_md):
        # Process only one lev for each
        if lcnt > 0: continue
        
        # Get q and spins for this simulation
        sim_lev_md = sim_md[lev]
        try: s1split = sim_lev_md['initial-dimensionless-spin1'].split()
        except:
            try: s1split = sim_lev_md['initial-spin1'].split()
            except: s1split = sim_lev_md['relaxed-spin1'].split()
        try: s2split = sim_lev_md['initial-dimensionless-spin2'].split()
        except:
            try: s2split = sim_lev_md['initial-spin2'].split()
            except: s2split = sim_lev_md['relaxed-spin2'].split()
        
        sim_s1x, sim_s1y, sim_s1z = np.array([float(val.strip(',')) for val in s1split])
        sim_s2x, sim_s2y, sim_s2z = np.array([float(val.strip(',')) for val in s2split])
        sim_q = float(sim_lev_md['initial-mass1']) / float(sim_lev_md['initial-mass2'])
        #print sim_q, sim_s1x, sim_s1y, sim_s1z, sim_s2x, sim_s2y, sim_s2z
        
        # Check if this simulations is not precessing?
        if not (abs(sim_s1x) < SEPS and abs(sim_s1y) < SEPS and abs(sim_s2x) < SEPS and abs(sim_s2y) < SEPS):
            #print "%s" % sim_lev_md['alternative-names'], " ignored due to transverse spins"
            ic1 += 1
            continue
        
        # Make sure NS spin = 0
        if abs(sim_s2z) > SEPS:
            #print "%s" % sim_lev_md['alternative-names'], " ignored due to spin on NS"
            ic2 += 1
            continue
            
        # Make sure BH spin != 0
        if abs(sim_s1z) < SEPS:
            #print "%s" % sim_lev_md['alternative-names'], " ignored due to NO spin on BH"
            ic3 += 1
            continue
        
        # Exclude a range of mass ratios
        q = sim_q
        q = max(q, 1./q)
        if q < 2.90 or q > 5.1:
            ic4 += 1
            continue
        
        print "Simulation: ", sim_lev_md['alternative-names']
        print "mass ratio = ", q
        print "spins of first = %.6f,%.6f,%.6f,\n\t second = %.6f,%.6f,%.6f" %\
                            (sim_s1x, sim_s1y, sim_s1z, sim_s2x, sim_s2y, sim_s2z)
        print "starting radial separation = ", sim_lev_md['initial-separation']
        print "number of orbits = ", sim_lev_md['number-of-orbits']
        print "time to merger = ", sim_lev_md['common-horizon-time']
        print "PERFECT MATCH!\n\n"

print "In order of conditions being checked:"
print "%d" % ic1, " ignored due to transverse spins"
print "%d" % ic2, " ignored due to spin on NS"
print "%d" % ic3, " ignored due to NO spin on BH"
print "%d" % ic4, " ignored due to mass-ratio"
```
